#### taks 3 fill missing data 
first make a judgment about what data are missing 
then figure out different ways of fill the missing data 

```python
#checking: for a given DataFrame, calculates how many values for 
    #each variable is null and prints the resulting table 
def print_null_freq(data):
    df = readdata(data)
    df_lng = pd.melt(df)
    null_variables = df_lng.value.isnull()
    return pd.crosstab(df_lng.variable, null_variables)
```

```python
print_null_freq('cs-training')
```

```python
#define functions to fill data with different methods
from sklearn.neighbors import KNeighborsRegressor
iv_imputer = KNeighborsRegressor(n_neighbors=1)
#define a function to split data into two categories: one with missing values, 
#and one without
def split_null(data,variable_name):
    df = readdata(data)
    global data_nonull 
    data_nonull = eval ('df[df.' + variable_name +'.isnull()==False]')
    global data_null
    data_null = eval ('df[df.' + variable_name + '.isnull()==True]')
    return data_nonull, data_null
```

```python
split_null('cs-training','MonthlyIncome')
```

```python
#define a function to fill missing data based on kneighborhood imputation of 
#the two most correlated variables 
def fill_kneighb_miss(var1, var2, var_to_fill):
    cols = [var1,var2]
    fill_var = data_nonull[var_to_fill]
    iv_imputer.fit(data_nonull[cols], fill_var)
    new_values = iv_imputer.predict(data_null[cols])
    data_null[var_to_fill]=new_values
    global newdata 
    newdata = data_nonull.append(data_null)
    return newdata
```

```python
import re

def camel_to_snake(column_name):
    """
    converts a string that is camelCase into snake_case
    Example:
        print camel_to_snake("javaLovesCamelCase")
        > java_loves_camel_case
    See Also:
        http://stackoverflow.com/questions/1175208/elegant-python-function-to-convert-camelcase-to-camel-case
    """
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', column_name)
    s2 = re.sub('-', '', s1)
    s3 = re.sub(': ', '_', s2)
    s4 = re.sub('0.1' , '0_1', s3 )
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s4).lower()
```

```python
#building previous functions into this 
#need to generate a feature (a column?)
def type_variable(data):
    df = readdata(data)
    df.columns = [camel_to_snake(col) for col in list(df.columns.values)]
    for i in range(len(df.columns)):
        var_name = str(list(df.columns.values)[i])
        command1 = 'type(list(df.' + var_name + '.unique())[0]) == str'
        command2 = 'len(list(df.' + var_name+ '.unique())) >= 4'
        if eval(command1)==False and eval(command2) :
            print (var_name, 'continuous')
        else:
            print (var_name,'categorical')
            
```

```python
#discretizing continuous variables 
def same_distance(var_name, binsnum):
    command = 'pd.cut(df.' + var_name +',bins=' + binsnum +')'
    result = eval (command)

    return pd.value_counts(result)    
```

```python
same_distance('age','5')
```

```python
#discretizing by quantile 
#qnum = number of buscket of equal size
def same_size(var_name, qnum):
    command = 'pd.qcut(df.' + var_name + ',' + qnum + ')'
    result = eval (command)
    return pd.value_counts(result)
```

```python
same_size('age','5')
```

```python
#get dummny variables from categorical variables 
def dummy_data(data, cate_var):
    df = readdata(data)
    df_dummy = pd.get_dummies(df[cate_var])
    df_new = df.join(df_dummy)
    return df_new
```
